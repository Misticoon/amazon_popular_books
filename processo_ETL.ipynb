```python
# Importar pacote
import pandas as pd
```

```python
# Extrair dados
livros_populares = pd.read_csv('Amazon_popular_books_dataset_tratado.csv')
```

```python
# Definir uma classe
class CatalogoLivros:
    def __init__(self, dataframe):
        self.catalogo = dataframe

    def adicionar_livro(self, novo_livro):
        # Adicionar uma nova linha ao DataFrame
        self.catalogo = self.catalogo.append(novo_livro, ignore_index=True)

    def excluir_livro_por_indice(self, indice):
        # Excluir uma linha do DataFrame com base no índice
        self.catalogo = self.catalogo.drop(index=indice).reset_index(drop=True)

    def listar_livros(self):
        # Listar informações de todos os livros no catálogo
        return self.catalogo
```

```python
# Criar catálogo de livros
catalogo = CatalogoLivros(livros_populares)
```

```python
# Excluir um livro por índice (por exemplo, o primeiro livro)
catalogo.excluir_livro_por_indice(0)
```

```python
# Listar todos os livros no catálogo
catalogo.listar_livros()
```

```python
# Adicionar um novo livro ao catálogo
novo_livro = {
    'ISBN10': '0007350813',
    'title': 'Wuthering Heights (Collins Classics)',
    'brand': 'Emily Brontë',
    'discount': 1.06,
    'initial_price': 7.99,
    'final_price': 6.93,
}

catalogo.adicionar_livro(novo_livro)
```

```python
# Listar todos os livros no catálogo após a exclusão
nova_arquivo = catalogo.listar_livros()
```

```python
# Salvar o DataFrame resultante no arquivo CSV
nova_arquivo.to_csv('Amazon_popular_books_dataset_processado.csv', index = False)
```
